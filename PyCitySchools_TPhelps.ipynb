```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Declare Variables/Calculations to use in DataFrame
schools = school_data["school_name"].nunique()
school_name = student_data[["school_name"]]
students = student_data["student_name"].count() 
total_budget = school_data["budget"].sum()
avg_math_score = student_data["math_score"].mean()
avg_reading_score = student_data["reading_score"].mean()
overall_passing_rate = ((avg_math_score + avg_reading_score)/2)

passing_math = student_data["math_score"] > 70
passing_math_count = passing_math[passing_math].count()
perc_passing_math = (passing_math_count / students)*100

passing_reading = student_data["reading_score"] > 70
passing_reading_count = passing_reading[passing_reading].count()
perc_passing_reading = (passing_reading_count / students)*100

#Make Dictionary of variables & column names to pass into DataFrame
raw_data = {'School count' : schools,
            "Total Students" : students,
               "Total Budget" : total_budget,
               'Avg Math' : avg_math_score,
               "Avg Reading" : avg_reading_score,
               "% Passing Math" : perc_passing_math,
               "% Passing Reading" : perc_passing_reading,
               "Overall passing" : [overall_passing_rate],}

#Display DF and format data.
new_school_df = pd.DataFrame(raw_data)
new_school_df["Total Students"] = new_school_df["Total Students"].map("{:,}".format)
new_school_df["Total Budget"] = new_school_df["Total Budget"].map("${:,}".format)
new_school_df.round(2)
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#merge data sets together to create merge_table
merge_table = pd.merge(student_data, school_data, on="school_name", how="outer")

#new DF as groupby from merge_table
schools_grouped_df = merge_table.groupby(['school_name'])

#declare variables
school_type = merge_table["type"].unique()
school_name = merge_table["school_name"]
total_students = schools_grouped_df["Student ID"].count()
total_school_budget = schools_grouped_df["budget"].unique()
per_student_budget = (total_school_budget / total_students)
avg_math_score = schools_grouped_df["math_score"].mean()
avg_reading_score = schools_grouped_df["reading_score"].mean()
passing_maths = merge_table[merge_table['math_score'] >=70].groupby(['school_name'])
passing_maths_count = passing_maths['math_score'].count()
passing_readings = merge_table[merge_table['reading_score'] >=70].groupby(['school_name'])
passing_readings_count = passing_readings['reading_score'].count()
percent_passing_math = (passing_maths_count / total_students)
percent_passing_reading = (passing_readings_count / total_students)
overall_passing_score = ((percent_passing_reading + percent_passing_math) / 2)
school_data = school_data.set_index("school_name")
school_type1 = school_data["type"]
total_school_names = schools_grouped_df["school_name"].unique()

#create DF
new_data_df = pd.DataFrame({
    "School Name" : total_school_names,
    "School Type" : school_type1,
    "Total Students" : total_students,
    "Total School Budget" : total_school_budget.astype(int),
    "Per Student Budget" : per_student_budget.astype(float),
    "Avg Math Score" : avg_math_score.round(2),
    "Avg Reading Score" : avg_reading_score.round(2),
    "% Passing Math" : percent_passing_math,
    "% Passing Reading" : percent_passing_reading,
    "Overall Passing Rate" : overall_passing_score})
#perform formating operations
new_data_df["Total School Budget"] = new_data_df["Total School Budget"].map("${:,}".format)
new_data_df["Per Student Budget"] = new_data_df["Per Student Budget"].map("${:,}".format)
new_data_df["% Passing Math"] = new_data_df["% Passing Math"].map('{:,.2%}'.format)
new_data_df["% Passing Reading"] = new_data_df["% Passing Reading"].map('{:,.2%}'.format)
new_data_df["Overall Passing Rate"] = new_data_df["Overall Passing Rate"].map('{:,.2%}'.format)

#set index
new_data_df = new_data_df.set_index("School Name")
#order DF
new_data_df = new_data_df.sort_values(["Overall Passing Rate"], ascending=False)

new_data_df.head()



```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#use 'loc' function to declare variables on grade value
ninth_grade_math = merge_table.loc[merge_table.grade == "9th", "math_score"]
tenth_grade_math = merge_table.loc[merge_table.grade == "10th", "math_score"]
elev_grade_math = merge_table.loc[merge_table.grade == "11th", "math_score"]
twel_grade_math = merge_table.loc[merge_table.grade == "12th", "math_score"]

#create series with variables & their index
ninth_grade_s = pd.Series(ninth_grade_math, index = ninth_grade_math.index)
tenth_grade_s = pd.Series(tenth_grade_math, index = tenth_grade_math.index)
elev_grade_s = pd.Series(elev_grade_math, index = elev_grade_math.index)
twel_grade_s = pd.Series(twel_grade_math, index = twel_grade_math.index)

#create series for school name master index to place grade series along
school_name_s = pd.Series(merge_table['school_name'], index=merge_table.index)

#create dataframe
grade_summary_math = pd.DataFrame({
    '9th Grade': ninth_grade_s,
    '10th Grade': tenth_grade_s,
    "11th Grade" : elev_grade_s,
    "12th Grade" : twel_grade_s})

#set index, groupby to display, and round
grade_summary_math = grade_summary_math.set_index(school_name_s)
grade_sum_math = grade_summary_math.groupby(["school_name"]).mean()
grade_sum_math.round(2)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Declare bin names and ranges
Spending_Ranges = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#create DF from existing variables to group into bin ranges
spending_df = pd.DataFrame({
    "Per Student Budget" : per_student_budget.astype(int),
    "Avg Math Score" : avg_math_score,
    "Avg Reading Score" : avg_reading_score,
    "% Passing Math" : percent_passing_math * 100,
    "% Passing Reading" : percent_passing_reading * 100,
    "% Overall Passing" : overall_passing_score * 100})
#cut DF into bins
spending_df["Spending_Ranges"] = pd.cut(spending_df["Per Student Budget"], Spending_Ranges, labels=group_names)

#groupby and reset bins to index
groupy_sort_sum = spending_df.groupby(["Spending_Ranges"]).mean().reset_index().groupby('Spending_Ranges').max()

#drop budget column used to cut DF into bins
final_df = groupy_sort_sum.drop(columns=['Per Student Budget'])
final_df                    
```
